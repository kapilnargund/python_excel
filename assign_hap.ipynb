```python
parents = list(set(df['Parent Company ID']))
parents = sorted([int(u[1:]) for u in parents])
list(map('b',parents))
```

```python
# What is the average total USD rev/parent?
```

```python
avg_total_li = []

for j in parents:
    df1 = df[df['Parent Company ID'] == j].reset_index(drop=True)

    sum1 = 0

    for o in df1.columns[-4:]:
        sum1 = sum1 + sum(df1[o])

    print('average total USD rev for', j,round(sum1/len(df1),2))
    avg_total_li.append(round(sum1/len(df1),2))
    
df_avg_total = pd.DataFrame()
df_avg_total['Parent Company ID'] = parents
df_avg_total['avg_total_rev'] = avg_total_li
df_avg_total
```
